# A4-Q5: Audio Enhancing

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.io.wavfile
from IPython.display import Audio
```

```python
# Load the audio recording
Omega, f = scipy.io.wavfile.read('recording.wav')
Audio(f, rate=Omega)
```

```python
# Some useful values
N = len(f)    # total number of samples
L = N / Omega # length of sound clip (in seconds)
t = np.arange(0,N) * L/N  # array of time stamps for samples
```

```python
# Corresponding array of sampled frequencies
omega = np.fft.fftshift(np.arange(-N/2, N/2)) / L
```

```python
plt.plot(t, f)
plt.title('Sound')
plt.xlabel('Time (seconds)')
plt.ylabel('Amplitude');
```

## (a)

```python
# do dft transform on f to get the modulus of the coefficients
F = np.fft.fft(f)
```

```python
plt.plot(omega, abs(F))
plt.title('Fourier Coeffcient Modulus vs. Frequency')
plt.xlabel('Frequency')
plt.ylabel('Fourier Coeffcient Modulus');
```

## (b)

#### We should get rid of the high-frequency range in the middle of the spectrum because the people's voice has much lower frequency than other noise, so the threshold may be set to 2200 according to the plot shown in (a).

```python
# threshold
Thr = 2200

# filter out the frequency whose magnitude is larger than Thr
n = 0  # the index of coefficients in F
while n <= N - 1:
    # If the frequency has a modulus greater than Thr, then set its coefficient to be zero
    if abs(omega[n]) > Thr:
        F[n] = 0
    n += 1
```

```python
plt.plot(omega, abs(F))
plt.title('Fourier Coeffcient Modulus vs. Frequency')
plt.xlabel('Frequency')
plt.ylabel('Fourier Coeffcient Modulus');
```

## (c)

```python
# do idft on F to get the sound signal
f_filtered = np.fft.ifft(F)
```

```python
# plot the  filtered sound (in the time domain)
plt.plot(t, f_filtered)
plt.title('Filtered Sound')
plt.xlabel('Time (seconds)')
plt.ylabel('Amplitude');
```

## (d)

```python
Audio(f_filtered, rate=Omega)
```
